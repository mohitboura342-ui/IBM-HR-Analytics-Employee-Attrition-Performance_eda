# Employee Attrition Analysis Using IBM HR Analytics Dataset

## Introduction

Employee attrition refers to the process through which employees leave an organization voluntarily or involuntarily. High employee attrition can negatively impact organizational performance by increasing recruitment and training costs, reducing productivity, and causing the loss of experienced employees and organizational knowledge. Therefore, understanding the factors that influence employee attrition is important for organizations seeking to improve employee retention and overall performance.

The IBM HR Analytics Employee Attrition & Performance dataset provides valuable employee-related information that can be used to analyze attrition patterns and identify the key factors associated with employee turnover. This project aims to explore the dataset through Exploratory Data Analysis (EDA) and investigate how various demographic, organizational, and job-related factors influence employee attrition.

# Literature Review  1

Employee attrition has been widely studied in human resource management and organizational behavior research. Previous studies suggest that employee turnover is influenced by multiple organizational, personal, and work-related factors.

Research by Mobley (1977) identified job satisfaction as one of the most important predictors of employee turnover. Employees who are dissatisfied with their jobs are more likely to seek alternative employment opportunities. Similarly, Abbasi and Hollman (2000) emphasized that compensation, recognition, organizational commitment, and management practices play significant roles in employee retention.

Arthur (1994) found that effective human resource management practices can reduce employee turnover and improve organizational performance. Walker (2001) highlighted the importance of employee development, career growth opportunities, and workplace support in retaining employees. Kim and Stoner (2008) reported that burnout, role stress, lack of autonomy, and insufficient social support contribute significantly to employee turnover intentions.

Other studies have identified poor work-life balance, excessive workload, lack of promotions, weak workplace relationships, and unsatisfactory work environments as major drivers of employee attrition. Employee demographics such as age, marital status, education, and work experience have also been shown to influence turnover behavior.

Overall, the literature suggests that employee attrition is a complex phenomenon resulting from the interaction of compensation-related, organizational, personal, and career-development factors.

## Removal of Irrelevant Columns

During the data cleaning phase, four variables were identified as non-informative and removed from the dataset before conducting Exploratory Data Analysis (EDA) and machine learning modeling.

### 1. EmployeeCount

**Description:**
The `EmployeeCount` variable represents the total employee count.

**Reason for Removal:**

* This column contains the same value (`1`) for every employee record.
* Since all observations have identical values, the variable has **zero variance**.
* Variables with zero variance do not provide any information that can help distinguish between employees who leave the organization and those who stay.
* Such columns do not contribute to statistical analysis, visualization, or predictive modeling.

**Conclusion:**
`EmployeeCount` was removed because it provides no useful information and does not influence employee attrition.

---

### 2. Over18

**Description:**
The `Over18` variable indicates whether an employee is over 18 years old.

**Reason for Removal:**

* Every employee in the dataset has the value `"Y"`.
* The variable contains only one category and therefore has no variation.
* Since there are no employees under the age of 18, this feature cannot explain differences in attrition behavior.
* Including such a column increases dataset dimensionality without adding meaningful information.

**Conclusion:**
`Over18` was removed because it contains only a single category and has no predictive value.

---

### 3. StandardHours

**Description:**
The `StandardHours` variable represents the standard working hours assigned to employees.

**Reason for Removal:**

* All employees have the same value (`80`) in this column.
* There is no variation across observations.
* A feature with a constant value cannot contribute to identifying patterns, relationships, or differences among employees.
* It does not provide any information regarding employee satisfaction, performance, or attrition.

**Conclusion:**
`StandardHours` was removed because it is a constant variable with no analytical significance.

---

### 4. EmployeeNumber

**Description:**
The `EmployeeNumber` variable is a unique identification number assigned to each employee.

**Reason for Removal:**/

* It serves only as an identifier and does not represent any employee characteristic.
* The assigned numbers are arbitrary and do not have a meaningful relationship with attrition.
* Including unique identifiers in machine learning models may introduce unnecessary noise.
* The model may attempt to learn patterns from employee IDs that have no real-world significance.
* Such variables can negatively affect model interpretability and performance.

**Conclusion:**
`EmployeeNumber` was removed because it is a unique identifier and does not contribute to understanding employee attrition.

# Univariate Analysis

## Exploring Employee Age Patterns Through Univariate Analysis

```python

plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='Age',
    bins=12,
    kde=True,
    color='royalblue',
    edgecolor='black'
)

plt.axvline(df['Age'].mean(),
            color='red',
            linestyle='--',
            linewidth=2,
            label=f"Mean Age: {df['Age'].mean():.1f}")

plt.axvline(df['Age'].median(),
            color='green',
            linestyle='-',
            linewidth=2,
            label=f"Median Age: {df['Age'].median():.1f}")

plt.title('Employee Age Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

### Findings

- The employee age distribution follows an approximately **normal (bell-shaped) distribution**, indicating a balanced workforce age structure.
- The **mean age (36.9 years)** and **median age (36 years)** are very close, suggesting that the data is nearly symmetrical with little skewness.
- Most employees are concentrated in the **30–40 years age range**, making it the largest employee group.
- There are fewer employees in the younger age group (**below 25 years**) and older age group (**above 50 years**).
- Since the mean and median are nearly equal, there is **no significant impact of outliers** on the age distribution.

### Findings:

- Among employees who work overtime, the attrition rate increases significantly to **30.5%**, with only **69.5%** staying.
- The attrition rate for overtime employees is nearly **3 times higher** than for employees who do not work overtime.
- This suggests that **OverTime is a strong factor influencing employee attrition**, and reducing excessive overtime may help improve employee retention.

```python
# Create Total Working Years Groups
df['TotalWorkingYearsGroup'] = pd.cut(
    df['TotalWorkingYears'],
    bins=[-1, 2, 5, 10, 20, 40],
    labels=['0-2 Years', '3-5 Years', '6-10 Years', '11-20 Years', '21+ Years']
)

# Calculate Attrition Rate
total_working_attrition = (
    df.groupby('TotalWorkingYearsGroup', observed=False)['Attrition']
    .apply(lambda x: (x == 'Yes').mean() * 100)
    .reset_index(name='Attrition Rate (%)')
)

# Bar Chart
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=total_working_attrition,
    x='TotalWorkingYearsGroup',
    y='Attrition Rate (%)',
    hue='TotalWorkingYearsGroup',
    palette='magma',
    legend=False
)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.1f}%',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.title('Attrition Rate by Total Working Years', fontsize=16, fontweight='bold')
plt.xlabel('Total Working Years Group', fontsize=12)
plt.ylabel('Attrition Rate (%)', fontsize=12)
plt.ylim(0, total_working_attrition['Attrition Rate (%)'].max() + 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

display(total_working_attrition)


# KDE Plot
plt.figure(figsize=(11, 6))

sns.kdeplot(
    data=df,
    x='TotalWorkingYears',
    hue='Attrition',
    fill=True,
    common_norm=False,
    alpha=0.35,
    palette='Set2'
)

plt.title('Total Working Years Density by Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Total Working Years', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


# ==================================================
# 2. Attrition vs Years At Company
# ==================================================

# Create Years at Company Groups
df['YearsAtCompanyGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=[-1, 1, 3, 5, 10, 20, 50],
    labels=['0-1 Year', '2-3 Years', '4-5 Years', '6-10 Years', '11-20 Years', '21+ Years']
)

# Calculate Attrition Rate
company_years_attrition = (
    df.groupby('YearsAtCompanyGroup', observed=False)['Attrition']
    .apply(lambda x: (x == 'Yes').mean() * 100)
    .reset_index(name='Attrition Rate (%)')
)

# Bar Chart
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=company_years_attrition,
    x='YearsAtCompanyGroup',
    y='Attrition Rate (%)',
    hue='YearsAtCompanyGroup',
    palette='viridis',
    legend=False
)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.1f}%',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.title('Attrition Rate by Years at Company', fontsize=16, fontweight='bold')
plt.xlabel('Years at Company Group', fontsize=12)
plt.ylabel('Attrition Rate (%)', fontsize=12)
plt.ylim(0, company_years_attrition['Attrition Rate (%)'].max() + 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

display(company_years_attrition)


# KDE Plot
plt.figure(figsize=(11, 6))

sns.kdeplot(
    data=df,
    x='YearsAtCompany',
    hue='Attrition',
    fill=True,
    common_norm=False,
    alpha=0.35,
    palette='Set3'
)

plt.title('Years at Company Density by Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Years at Company', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

| Feature | TotalWorkingYears | YearsAtCompany |
|---|---|---|
| Definition | Total number of years an employee has worked throughout their entire career. | Number of years an employee has worked in the current company only. |
| Includes | Experience from current and previous companies. | Experience from the current company only. |
| Purpose in EDA | Helps analyze whether overall work experience affects employee attrition. | Helps analyze whether employee tenure and loyalty affect attrition. |
| Example | An employee worked 6 years in previous companies and 4 years in the current company: TotalWorkingYears = 10. | For the same employee: YearsAtCompany = 4. |
| Interpretation | Represents the employee’s overall career experience. | Represents the employee’s duration or tenure in the organization. |

### Findings

- Employees with **0–2 total working years** have the highest attrition rate at **43.9%**, indicating that early-career employees are more likely to leave.  
- Attrition decreases consistently as total work experience increases, reaching only **7.7%** for employees with **21+ years** of experience.  
- The density plot shows that employees who left the company are mostly concentrated in the lower experience range, especially below **10 total working years**.  
- Employees with **0–1 year at the company** have the highest attrition rate of **34.9%**, showing that the first year is a critical retention period.  
- Attrition generally declines as years at the company increase, with the lowest rate of **6.7%** among employees with **11–20 years** of tenure.  
- Overall, both lower total work experience and shorter company tenure are strongly associated with higher employee attrition.
